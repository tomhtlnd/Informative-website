```python
df_fs = pd.read_csv('data/fossil_fuel_support.csv')
df_fs.head()
```

```python
df_fs = df_fs[(df_fs['Support Mechanism'] == 'Total') & (df_fs['Beneficiary or Sector'] == 'Total')]
df_fs = df_fs.drop(['COUNTRY', 'MECHANISM', 'INDIC_OR_SEC', 'Beneficiary or Sector', 'PROD', 'MEASURE', 'YEAR', 'Flag Codes', 'Flags'], axis=1)
df_fs['Country'].replace({'China (People\'s Republic of)': 'China'}, inplace=True)

df_dollar = df_fs[df_fs['Unit'] == 'National currency (in millions)']
df_gdp = df_fs[df_fs['Unit'] == 'as % of GDP']

#ff_energy.rename({'Energy_Type': 'Energy Type'}, axis=1, inplace=True)
#df_energy['Technology'].replace({'Hydropower (excl. Pumped Storage)': 'Hydropower', "Other non-renewable energy": "Other non-renewable"}, inplace=True)
#df_energy['Country'].replace({'China, P.R.: Mainland': 'China', 'Korea, Rep. of': 'Korea', 'Russian Federation': 'Russia' }, inplace=True)

df_gdp.head()
```

```python
top10 = ['China','United States','India','Russia','Japan','Canada','Brazil','Korea','Germany','France', 'United Kingdom']
df_dollar_top10 = df_dollar[df_dollar['Country'].isin(top10)]
df_dollar_top10 = df_dollar_top10[df_dollar_top10['Fuel Type'] == 'Total']
df_dollar_top10 = df_dollar_top10.drop(['Unit'], axis=1)
df_dollar_top10
```

```python
fig = px.line(df_dollar, 
              x='Year', 
              y='Value',  
              title='Fossil fuel support by country',
              color='Country',
              height=600,width=800)
fig.update_yaxes(
     title_text = "National currency (in millions)")
fig.show()
py.plot(fig, filename = 'Fossil fuel support', auto_open=False)
```

```python
g20 = df_fs[df_fs['Country'] == 'G20 and EaP partner economies']
g20
```

```python
g20_dollar = g20[(g20['Unit'] == 'US Dollars (nominal) [in millions]') & (g20['Fuel Type'] != 'Total')]
g20_dollar = g20_dollar.groupby(['Year', 'Fuel Type'])['Value'].sum()
g20_dollar = g20_dollar.reset_index()
g20_dollar
```

```python
fig = px.bar(g20_dollar, 
              x='Year', 
              y='Value',  
              title='Fossil fuel support by country',
              color='Fuel Type',
              height=600,width=800,
              category_orders={'Fuel Type': g20_dollar.sort_values('Value', ascending=False)['Fuel Type'].unique()},
              color_discrete_map={'Petroleum': px.colors.qualitative.Prism[6], 'End-use electrictiy': px.colors.qualitative.Prism[2], 'Natural gas': px.colors.qualitative.Prism[1], 'Coal': 'black'})
fig.show()
```
